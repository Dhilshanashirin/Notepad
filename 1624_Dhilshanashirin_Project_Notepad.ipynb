```python
import tkinter as tk
from tkinter import filedialog, Text

# Create the main application window
root = tk.Tk()
root.title("Simple Notepad")

# Create a Text widget for text editing
text_area = Text(root, undo=True)
text_area.pack(fill="both", expand=True)

# Define functions for file operations
def new_file():
    text_area.delete(1.0, tk.END)
    root.title("Untitled - Simple Notepad")

def open_file():
    file_path = filedialog.askopenfilename(defaultextension=".txt",
                                           filetypes=[("Text Files", "*.txt"), 
                                                      ("All Files", "*.*")])
    if file_path:
        with open(file_path, "r") as file:
            text_area.delete(1.0, tk.END)
            text_area.insert(tk.END, file.read())
        root.title(f"{file_path} - Simple Notepad")

def save_file():
    file_path = filedialog.asksaveasfilename(defaultextension=".txt",
                                             filetypes=[("Text Files", "*.txt"), 
                                                        ("All Files", "*.*")])
    if file_path:
        with open(file_path, "w") as file:
            file.write(text_area.get(1.0, tk.END))
        root.title(f"{file_path} - Simple Notepad")

# Create a menu bar
menu_bar = tk.Menu(root)

# File menu
file_menu = tk.Menu(menu_bar, tearoff=0)
file_menu.add_command(label="New", command=new_file)
file_menu.add_command(label="Open", command=open_file)
file_menu.add_command(label="Save", command=save_file)
file_menu.add_separator()
file_menu.add_command(label="Exit", command=root.quit)

menu_bar.add_cascade(label="File", menu=file_menu)

# Configure the application to display the menu
root.config(menu=menu_bar)

# Run the application
root.mainloop()
```

```python
from tkinter import *
from tkinter import filedialog

window = Tk()
window.title('Notepad')

class Notepad:
    def __init__(self, window):
        self.window = window
        self.window.title("Untitled - Notepad")
        
        
        self.text_area = Text(window, wrap='word', undo=True)
        self.text_area.pack(expand=True, fill=BOTH)
        
        
        self.menu_bar = Menu(window)
        
        
        self.file_menu = Menu(self.menu_bar, tearoff=0)
        self.file_menu.add_command(label="New", command=self.new_file)
        self.file_menu.add_command(label="Open", command=self.open_file)
        self.file_menu.add_command(label="Save", command=self.save_file)
        self.file_menu.add_separator()
        self.file_menu.add_command(label="Exit", command=window.quit)
        self.menu_bar.add_cascade(label="File", menu=self.file_menu)
        
        self.window.config(menu=self.menu_bar)

    def new_file(self):
        self.text_area.delete(1.0, END)
        self.window.title("Untitled - Notepad")

    def open_file(self):
        file_path = filedialog.askopenfilename(defaultextension=".txt",
                                               filetypes=[("Text Files", "*.txt"),
                                                          ("All Files", "*.*")])
        if file_path:
            with open(file_path, "r") as file:
                self.text_area.delete(1.0, END)
                self.text_area.insert(END, file.read())
            self.window.title(f"{file_path} - Notepad")

    def save_file(self):
        file_path = filedialog.asksaveasfilename(defaultextension=".txt",
                                                 filetypes=[("Text Files", "*.txt"),
                                                            ("All Files", "*.*")])
        if file_path:
            with open(file_path, "w") as file:
                file.write(self.text_area.get(1.0, END))
            self.window.title(f"{file_path} - Notepad")
    def exit_file(self):
        file_path = filedialog.askexitfilename(defaultextension="txt"),
                                                filetypes=[("Text Files", "*.txt"),
                                                          ("All Files", "*.*")]

notepad_app = Notepad(window)

window.mainloop()
```
